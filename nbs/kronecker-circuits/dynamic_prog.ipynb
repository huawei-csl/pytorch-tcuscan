# Dynamic programming for depth-size optimization

```python
from functools import cache
from math import ceil


@cache
def kron_circuit(n: int, d: int) -> int:
  "Returns the size, given length and depth"
  print(f"Call n: {n} | d : {d}")
  if n == 2 and d == 1:
    # Prefix sum of
    return 1
  if n == 3 and d == 2:
    return 2
  if n == 4 and d == 2:
    return 5
  if n > 2 and d == 1:
    print(f"[Requested] Impossible circuit with length {n} and depth {d}.")
    return 1e6
  if n == 2 and d < 1:
    return 1e6
  if n <= 0:
    return 0

  n_prime = ceil(n / 2)

  min_size = 2 * n_prime - 1 + kron_circuit(n_prime - 1, d - 2)

  # Keep track of size of smaller depth circuits
  for min_depth in range(2, d):
    current_size = kron_circuit(n, min_depth)
    if min_size is None or current_size < min_size:
      print(f"Used smaller depth: {min_depth}")
      min_size = current_size

  return min_size
```
