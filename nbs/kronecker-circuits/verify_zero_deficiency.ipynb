# Verify zero-deficiency

This notebook verifies for every combination of n and s that the Kronecker prefix circuits have indeed zero deficiency.
- Specifically, we there are two functions `get_depth(n, s)` and `get_size(n,s)` which return the depth and size of the circuit for given `n` and `s`
- Then, it tests if the zero-deficiency formula is satisfied: `get_size(n, s) + get_depth(n, s) == 2 * n - 2`

```python
import numpy as np
from matplotlib import pyplot as plt

def get_depth(_n,_s):
    if _n <= _s:
        return _n-1
    else:
        return _s + get_depth(int(np.ceil(_n/_s))-1, _s)

def get_size(_n, _s):
    if _n <= _s:
        return _n - 1
    else:
        _n_prime = int(np.ceil(_n/_s)) - 1
        sz = _n - _n_prime - 1 # first layer
        sz = sz + get_size(_n_prime, _s) # recursive layer
        sz = sz + _n - _s - (_n // _s - 1) # last layer
        if _n % s == 0:
            sz = sz + 1
        return sz
    
max_n = 2043
for s in range(2,17):
    for n in range(2,max_n+1):
        D = get_depth(n, s)
        S = get_size(n, s)
        # print(f'n = {n},    D + S = {D + S},     2n-2 = {2 * n - 2}')
        assert D + S == 2 * n - 2
print('Test passed')
```
