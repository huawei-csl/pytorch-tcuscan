```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from glob import glob
from os.path import join
```

```python
spmv_df = pd.concat(map(pd.read_csv, glob(join('../../', f"bench_results_spmv_128*.csv"))))
```

```python
diff_df = pd.concat(map(pd.read_csv, glob(join('../../', f"bench_results_diff*.csv"))))
```

```python
mcscan_df = pd.concat(map(pd.read_csv, glob(join('../../', f"bench_results_mcscan*.csv"))))
```

```python
gather_spmv_df = pd.concat(map(pd.read_csv, glob(join('../../', f"bench_results_gather_spmv*.csv"))))
```

```python
csr_gather_df = pd.concat(map(pd.read_csv, glob(join('../../', f"bench_results_csr_gather*.csv"))))
```

```python
tot_df = pd.concat([csr_gather_df, mcscan_df, gather_spmv_df,diff_df])
```

```python


tot_df_merged = pd.merge(csr_gather_df[['benchname', 'time_us']], mcscan_df[['benchname', 'time_us']], on='benchname', suffixes=('_csr_gather', '_mcscan'))
tot_df_merged = pd.merge(tot_df_merged, gather_spmv_df[['benchname', 'time_us']], on='benchname')
tot_df_merged = pd.merge(tot_df_merged, diff_df[['benchname', 'time_us']], on='benchname', suffixes=('_gather_spmv', '_diff'))
tot_df_merged = tot_df_merged.sort_values(by='benchname')
```

```python
tot_df_merged.columns = ['benchname', 'csr_gather', 'mcscan', 'gather_spmv', 'diff']
tot_df_merged.set_index('benchname', inplace=True)
tot_df_merged.plot(kind='bar', stacked=True, figsize=(10, 6), edgecolor='black')

plt.title('SpMV execution time breakdown')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Time [us]')
plt.show()
```

```python
spmv_df = pd.concat(map(pd.read_csv, glob(join('../../', f"bench_results_spmv_128*.csv"))))

spmv_df = spmv_df[spmv_df['operator'] == 'spmv_128']

spmv_merged = spmv_df[['benchname', 'time_us']]
spmv_merged.columns = ['benchname', 'spmv']

spmv_merged.set_index('benchname', inplace=True)
spmv_merged = spmv_merged.sort_values(by='benchname')

fig, ax = plt.subplots(figsize=(12, 6))

tot_df_merged.plot(kind='bar', stacked=True, ax=ax, position=1, width=0.4, edgecolor='black')

spmv_merged.plot(kind='bar', ax=ax, position=0, width=0.4, color='gray', legend=True, edgecolor = "black")

plt.title('SpMV execution time breakdown')
plt.xlabel('')
plt.xlim(-0.5, None)
plt.ylabel('Time [us]')
plt.xticks(rotation=45, ha='right')

plt.show()
```
