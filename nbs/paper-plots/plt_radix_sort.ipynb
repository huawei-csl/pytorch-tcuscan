```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from glob import glob
from os.path import join
```

```python
pd.set_option('display.max_rows', 1000)
radixsort_df = pd.concat(map(pd.read_csv, glob(join('../..', f"*sort*.csv"))))

radixsort_df['mem_bytes'] = (radixsort_df['size'] * 2)  + radixsort_df['outputsize'] * 2 + radixsort_df['outputsize'] * 2
radixsort_df['bandwidth_gbps'] = (radixsort_df['mem_bytes']) / (radixsort_df['time_us']*1000) #GB/s
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x='size', y='bandwidth_gbps', hue='operator', data=radixsort_df, marker='o')
```

```python
radixsort_df['time_ms'] = radixsort_df['time_us'] / 1000
plt.figure(figsize=(8, 5))
sns.lineplot(x='size', y='time_ms', hue='operator', data=radixsort_df, marker='o')
```
