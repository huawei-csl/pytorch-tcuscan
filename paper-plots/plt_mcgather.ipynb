```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from glob import glob
from os.path import join
```

```python
pd.set_option('display.max_rows', 1000)
gather_df = pd.concat(map(pd.read_csv, glob(join('..', f"*gather*.csv"))))

gather_df['mem_bytes'] = (gather_df['size'] * (4))  + gather_df['outputsize'] * 4 + gather_df['outputsize'] * 4
gather_df['bandwidth_gbps'] = (gather_df['mem_bytes']) / (gather_df['time_us']*1000) #GB/s
```

```python
gather_df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x='size', y='bandwidth_gbps', hue='operator', data=gather_df, marker='o')
```
