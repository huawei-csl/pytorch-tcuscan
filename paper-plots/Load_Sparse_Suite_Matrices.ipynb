## Example usage of ssgetpy

```python
from tqdm.notebook import tqdm
```

```python
from scipy.io import mmread
from scipy.sparse import csr_matrix

def convert_to_segments(A_ssget):
  "Converts a sparse matrix from ssget into a segmented sum/scan input (x, f)"

  file_location, _ = A_ssget.download(extract=True)
  name = A_ssget.name

  A = mmread(f"{file_location}/{name}.mtx")
  B = csr_matrix(A)

  # Data vector
  x = B.data

  # Flags vector
  f = np.zeros(B.nnz + 1)
  f[B.indptr] = 1

  return x, f
```

```python
# matrices = ssgetpy.search(nzbounds=(1000,1000000), isspd=False, limit=20)
matrices = ssgetpy.search(nzbounds=(100000,10000000), limit=200)

spmat_dict = {}
for mat in tqdm(matrices):
  name = mat.name
  x, f = convert_to_segments(mat)
  ratio = sum(f) / len(f)
  print(type(mat))
  print(f"Converting matrix {name} | rows {mat.rows} | cols {mat.cols} | nnzs: {mat.nnz} | ratio : {ratio}")
  spmat_dict[name] = (x, f)
```
